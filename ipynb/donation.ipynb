# donation

```python
with MyMongo() as db:
    donation = db.get_df_from_table('assembly', 'donation')
print(len(donation))
donation = donation.drop(columns=['_id', '연번']); donation
```

## filtered_in: x.num != Nan

```python
criteria = donation.isna().any(axis=1)
filtered_out = donation.loc[criteria]; filtered_out
filtered_in = donation.loc[(criteria.map(lambda x: not x))]; filtered_in
print(len(donation), '=?', len(filtered_out), '+', len(filtered_in))
```

# grb_don_by_com

```python
grb_don_by_com = com_mem_don.groupby('committeename').agg({'후원금 모금액': ['sum', 'count']}); grb_don_by_com
grb_don_by_com.columns = [' '.join(col).strip() for col in grb_don_by_com.columns.values]; grb_don_by_com
grb_don_by_com = grb_don_by_com.reset_index(); grb_don_by_com
grb_don_by_com = grb_don_by_com.rename(columns={'committeename':'상임위&상설특위','후원금 모금액 sum':'후원금합계',
                                        '후원금 모금액 count':'인원'}); grb_don_by_com
grb_don_by_com['후원금평균'] = grb_don_by_com['후원금합계'] / grb_don_by_com['인원']; grb_don_by_com
grb_don_by_com.sort_values('후원금평균', ascending=False)
```

```python
view_don_by_com = grb_don_by_com.rename(columns={'committeename':'상임위&상설특위','sum':'합계','count':'인원수','average':'평균'})
```
